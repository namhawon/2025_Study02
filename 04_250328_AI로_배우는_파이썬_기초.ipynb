# 파이썬 예외 처리
파이썬에서 예외 처리는 프로그램 실행 중 발생할 수 있는 오류를 처리하는 방법입니다. 예외가 발생하면 프로그램이 비정상적으로 종료될 수 있지만, 예외 처리를 통해 오류를 감지하고 적절한 조치를 취하여 프로그램이 계속 실행될 수 있도록 할 수 있습니다.



try:
    # 예외가 발생할 수 있는 코드
except 예외 as e:
    # 예외 발생 시 실행할 코드
finally:
    # 예외 발생 여부와 관계없이 항상 실행할 코드
    
- try 블록에는 예외가 발생할 수 있는 코드를 작성합니다.
- except 블록에는 예외 발생 시 실행할 코드를 작성합니다. 특정 예외를 지정하거나, 모든 예외를 처리할 수 있습니다.
- finally 블록에는 예외 발생 여부와 관계없이 항상 실행할 코드를 작성합니다. 주로 자원 정리 등에 사용됩니다.

# ValueError
ValueError는 파이썬에서 함수에 잘못된 유형의 인수가 전달되었지만 해당 유형이 그 함수에 적합하지 않을 때 발생하는 예외입니다. 즉, 함수가 예상하는 유형의 값이 아니거나 값의 범위가 적절하지 않을 때 발생합니다.

```python
# 1. 잘못된 인수 유형:정수를 기대하는 함수에 문자열 전달
def to_int(value):
  return int(value)

try:
  result = to_int("hello")
except ValueError as e:
  print(f"ValueError 발생: {e}") # ValueError 발생: invalid literal for int() with base 10: 'hello'
```

```python
# 2. 잘못된 값 범위:양수를 기대하는 함수에 음수 전달
import math

def sqrt_positive(value):
  if value < 0:
    raise ValueError("양수를 입력해야 합니다.")
  return math.sqrt(value)

try:
  result = sqrt_positive(-1)
except ValueError as e:
  print(f"ValueError 발생: {e}") # ValueError 발생: 양수를 입력해야 합니다.
```

```python
# 3. 잘못된 데이터 형식:날짜 형식이 잘못된 문자열을 날짜 객체로 변환
from datetime import datetime

def to_date(date_str):
  return datetime.strptime(date_str, "%Y-%m-%d")

try:
  result = to_date("2023-13-32")
except ValueError as e:
  print(f"ValueError 발생: {e}") # ValueError 발생: time data '2023-13-32' does not match format '%Y-%m-%d'
```

```python
# 4. 빈 시퀀스에서 요소 가져오기:빈 리스트에서 요소를 가져오려는 경우
my_list = []

try:
  item = my_list.pop()
except ValueError as e:
  print(f"ValueError 발생: {e}") # ValueError 발생: pop from empty list
```

```python
# 5. 잘못된 인코딩으로 문자열 디코딩:잘못된 인코딩으로 문자열 디코딩
encoded_string = "안녕하세요".encode("utf-8")

try:
  decoded_string = encoded_string.decode("latin-1")
except ValueError as e:
  print(f"ValueError 발생: {e}") # ValueError 발생: decode() input doesn't match the encoding scheme
```

# 문제 1: 두 수 비교하기

CoT:

*   사용자로부터 두 개의 정수를 입력받습니다.
*   두 수를 비교하여 어떤 수가 더 큰지, 또는 같은지 판별해야 합니다.
*   첫 번째 수가 두 번째 수보다 큰 경우를 if 문으로 확인합니다.
*   첫 번째 수가 두 번째 수보다 작은 경우를 elif 문으로 확인합니다.
*   위의 두 조건이 모두 거짓인 경우 else 문으로 두 수가 같음을 처리합니다.


```python
# 문제 3: 두 수 비교하기
num1_str = input("첫 번째 정수를 입력하세요: ")
num2_str = input("두 번째 정수를 입력하세요: ")

try:
    num1 = int(num1_str)
    num2 = int(num2_str)

    if num1 > num2:
        print(f"{num1}은(는) {num2}보다 큽니다.")
    elif num1 < num2:
        print(f"{num1}은(는) {num2}보다 작습니다.")
    else:
        print(f"{num1}과(와) {num2}은(는) 같습니다.")

except ValueError:
    print("유효한 정수를 입력해주세요.")
```

# 문제 2: 세 수 중 가장 큰 수 찾기

CoT:



*   사용자로부터 세 개의 정수를 입력받습니다. 각각 num1, num2, num3 변수에 저장합니다.
*   가장 큰 수를 저장할 변수 max_num 을 초기화합니다. 처음에는 첫 번째 수를 가장 큰 수라고 가정합니다
*   if 문을 사용하여 num2 가 현재 max_num 보다 큰지 비교합니다. 만약 크다면 max_num 을 num2 로 업데이트합니다.
*   다시 if 문을 사용하여 num3 가 현재 max_num 보다 큰지 비교합니다. 만약 크다면 max_num 을 num3로 업데이트합니다.
*   모든 비교가 끝나면 max_num 변수에 세 수 중 가장 큰 수가 저장됩니다.
*   최종적으로 max_num 값을 출력하여 결과를 사용자에게 보여줍니다.
*   입력 과정에서 발생할 수 있는 ValueError를 처리하여 사용자가 유효한 정수를 입력하도록 안내합니다.


```python
# 문제: 세 수 중 가장 큰 수 찾기
try:
    num1_str = input("첫 번째 정수를 입력하세요: ")
    num1 = int(num1_str)
    print(f"첫 번째 수: {num1}")

    num2_str = input("두 번째 정수를 입력하세요: ")
    num2 = int(num2_str)
    print(f"두 번째 수: {num2}")

    num3_str = input("세 번째 정수를 입력하세요: ")
    num3 = int(num3_str)
    print(f"세 번째 수: {num3}")

    max_num = num1  # 처음에는 num1이 가장 크다고 가정

    if num2 > max_num:
        print(f"두 번째 수 {num2}가 현재 최대값 {max_num}보다 큽니다. 최대값을 업데이트합니다.")
        max_num = num2
    else:
        print(f"두 번째 수 {num2}는 현재 최대값 {max_num}보다 크지 않습니다.")

    if num3 > max_num:
        print(f"세 번째 수 {num3}가 현재 최대값 {max_num}보다 큽니다. 최대값을 업데이트합니다.")
        max_num = num3
    else:
        print(f"세 번째 수 {num3}는 현재 최대값 {max_num}보다 크지 않습니다.")

    print(f"\n입력하신 세 수 {num1}, {num2}, {num3} 중에서 가장 큰 수는 {max_num}입니다.")

except ValueError:
    print("유효한 정수를 입력해주세요.")
```

```python
# 문제: 세 수 중 가장 큰 수 찾기

try:
    num1 = int(input("첫 번째 정수를 입력하세요: "))
    num2 = int(input("두 번째 정수를 입력하세요: "))
    num3 = int(input("세 번째 정수를 입력하세요: "))

    max_num = num1  # 처음에는 num1이 가장 크다고 가정

    if num2 > max_num:
        max_num = num2
    if num3 > max_num:
        max_num = num3

    print(f"\n입력하신 세 수 {num1}, {num2}, {num3} 중에서 가장 큰 수는 {max_num}입니다.")

except ValueError:
    print("유효한 정수를 입력해주세요.")
```

해당 코드는 세 개의 정수 중 가장 큰 수를 찾기 위한 프로그램이지만, if 다음에 elif를 사용함으로써 논리적인 오류가 발생한다. if, elif, else 구문은 조건들 중 하나만 실행되도록 설계된 구조이기 때문에, 첫 번째 if 조건이 참이면 그 이후의 elif 조건은 검사하지 않고 건너뛰게 된다.이 구조는 여러 조건 중 하나만 선택해야 하는 경우에는 적합하지만, 각각의 조건을 모두 검사해야 하는 상황에서는 부적절하다. 예를 들어, num2 > max_num 조건이 참일 경우 max_num은 num2로 업데이트되지만, 그 이후 elif num3 > max_num 조건은 실행되지 않으므로, num3이 가장 큰 수임에도 불구하고 반영되지 않는 오류가 발생할 수 있다.이러한 논리 오류를 방지하기 위해서는 elif가 아닌 두 개의 독립적인 if문을 사용하여 각각의 조건을 따로 평가해야 한다. 이렇게 하면 num2와 num3을 모두 비교할 수 있어 올바른 최대값을 찾을 수 있다.

# 문제 3: 학점 판별하기

CoT:



*   사용자로부터 시험 점수 (0부터 100 사이의 정수)를 입력받습니다.
*   입력받은 점수에 따라 학점을 판별해야 합니다.
*   90점 이상이면 'A' 학점입니다.
*   80점 이상 90점 미만이면 'B' 학점입니다.
*   70점 이상 80점 미만이면 'C' 학점입니다.
*   60점 이상 70점 미만이면 'D' 학점입니다.
*   60점 미만이면 'F' 학점입니다.
*   if, elif, else 문을 사용하여 점수 범위에 따라 학점을 결정하고 출력합니다.
*   입력 유효성 검사 (0~100 사이의 정수)를 추가합니다.

```python
# 문제 3: 학점 판별하기

score_str = input("시험 점수 (0~100)을 입력하세요: ")

try:
    score = int(score_str)
    if 0 <= score <= 100:
        if score >= 90:
            grade = 'A'
        elif score >= 80:
            grade = 'B'
        elif score >= 70:
            grade = 'C'
        elif score >= 60:
            grade = 'D'
        else:
            grade = 'F'
        print(f"당신의 학점은 {grade}입니다.")
    else:
        print("점수는 0부터 100 사이의 값이어야 합니다.")
except ValueError:
    print("유효한 정수를 입력하세요.")
```

# 문제 4: 윤년 판별하기

CoT:



*   사용자로부터 연도를 입력받습니다.
*   입력받은 연도가 윤년인지 아닌지 판별해야 합니다.
*   윤년의 조건은 다음과 같습니다.
*   4로 나누어 떨어지면서 100으로 나누어 떨어지지 않거나,400으로 나누어 떨어지는 해는 윤년입니다.
*   if 문과 논리 연산자 (and, or, not )를 사용하여 윤년 조건을 확인합니다.
*   조건에 따라 윤년인지 아닌지 출력합니다.


```python
# 문제 4: 윤년 판별하기

year_str = input("연도를 입력하세요: ")
try:
    year = int(year_str)
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        print(f"{year}년은 윤년입니다.")
    else:
        print(f"{year}년은 윤년이 아닙니다.")
except ValueError:
    print("유효한 정수를 입력하세요.")
```

# 문제 1: 1부터 N까지의 합 계산하기

CoT:

*   사용자로부터 양의 정수 N을 입력받습니다.
*   1부터 N까지의 모든 정수를 더해야 합니다.
*   합계를 저장할 변수를 초기화합니다(0으로 시작).
*   for 루프를 사용하여 1부터 N까지 순회합니다.
*   각 숫자를 합계 변수에 더합니다.
*   최종 합계를 출력합니다.


```python
# 문제 1: 1부터 N까지의 합 계산하기

n_str = input("양의 정수 N을 입력하세요: ")

try:
    n = int(n_str)
    if n <= 0:
        print("N은 양의 정수여야 합니다.")
    else:
        total = 0
        for i in range(1, n + 1):
            total += i
        print(f"1부터 {n}까지의 합은: {total} 입니다.")
except ValueError:
    print("유효한 정수를 입력하세요.")
```

두 코드의 차이는 print() 함수의 들여쓰기 위치에 있습니다. 이로 인해 실행 결과가 다르게 나타납니다.첫 번째 셀에서는 print(f"1부터 {n}까지의 합은: {total} 입니다.")가 for 반복문 밖에 있기 때문에, 반복문이 모두 끝난 후에 한 번만 실행되어 최종 합계만 출력됩니다. 반면 두 번째 셀에서는 print() 문이 for 반복문 안에 들여쓰기 되어 있어서, 반복이 실행될 때마다 매번 중간 결과를 출력하게 됩니다. 그 결과, 1부터 N까지 각 단계에서 누적된 합이 모두 출력됩니다.이처럼 파이썬에서는 들여쓰기가 코드의 실행 범위를 결정하므로, 같은 코드라도 들여쓰기 위치에 따라 실행 횟수와 출력 결과가 달라질 수 있습니다. 따라서 print()와 같은 출력문의 위치는 목적에 맞게 정확하게 들여써야 합니다.

# 문제 2: 구구단 출력하기 (특정 단)

CoT:

*   사용자로부터 출력하고 싶은 구구단의 단 (2부터 9 사이의 정수)을 입력받습니다.
*   입력된 단에 대해 1부터 9까지 곱한 결과를 출력해야 합니다.
*   for 루프를 사용하여 1부터 9까지 순회합니다.
*   각 숫자에 입력받은 단을 곱하고 결과를 "단 x숫자 = 결과" 형태로 출력합니다.
*   입력 유효성 검사 (2~9 사이의 정수)를 추가합니다.


```python
# 문제 2: 구구단 출력하기 (특정 단)

dan_str = input("출력하고 싶은 구구단의 단 (2~9)을 입력하세요: ")

try:
    dan = int(dan_str)
    if 2 <= dan <= 9:
        print(f"\n--- {dan}단 ---")
        for i in range(1, 10):
            result = dan * i
            print(f"{dan} x {i} = {result}")
    else:
        print("2부터 9 사이의 정수를 입력해주세요.")
except ValueError:
    print("유효한 정수를 입력하세요.")
```

# 문제 4: 특정 조건까지 반복하기 (while 루프)

CoT:

*   초기값 0부터 시작하는 변수를 하나 설정합니다.
*   while 루프를 사용하여 변수가 10보다 작을 동안 반복합니다.
*   루프 안에서 현재 변수의 값을 출력합니다.
*   루프가 반복될 때마다 변수의 값을 1씩 증가시킵니다.
*   변수가 10이 되면 루프가 종료됩니다.


```python
# 문제 4: 특정 조건까지 반복하기 (while 루프)

count = 0
print("0부터 9까지 출력:")

while count < 10:
    print(count)
    count += 1

print("while 루프 종료")
```

```python
# 문제 4: 특정 조건까지 반복하기 (for 루프)

print("0부터 9까지 출력:")

for count in range(10):
    print(count)

print("for 루프 종료")
```
