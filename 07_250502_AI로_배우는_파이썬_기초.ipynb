**1) 꺾은선 그래프**

## 7주차 강의노트 연습문제

1. 다음 데이터를 이용하여 1년 동안의 월별 평균 강수량을 꺾은선 그래프로 그려보세요.

- 월: ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

- 강수량(mm): [30, 40, 60, 80, 100, 150, 300, 280, 120, 70, 50, 35]

- 그래프 제목과 x, y축 레이블을 추가하세요.

```python
# 1) 월별 강수량 꺾은선 그래프
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
rainfall = [30, 40, 60, 80, 100, 150, 300, 280, 120, 70, 50, 35]

plt.figure(figsize=(8, 4))
sns.lineplot(x=months, y=rainfall, marker='o')
plt.title('Monthly Average Rainfall')
plt.xlabel('Month')
plt.ylabel('Rainfall (mm)')
plt.grid(True)
plt.show()
```

```python
# 3) 산점도(Scatter Plot)
kor = [547, 546, 554, 524, 526]
usa = [474, 487, 481, 470, 478]

plt.figure(figsize=(6,4))
plt.scatter(kor, usa)
plt.title('Korea vs USA Scores')
plt.xlabel('Korea Score')
plt.ylabel('USA Score')
plt.grid(True)
plt.show()
```

```python
# 9) Korea vs USA 꺾은선 비교 (각각)
year = [2006, 2009, 2012, 2015, 2018]
kor  = [547, 546, 554, 524, 526]
usa  = [474, 487, 481, 470, 478]

# Korea
plt.figure(figsize=(5,3))
plt.plot(year, kor, 'r--', label='Korea')
plt.title('Korea Mathematics Performance')
plt.xlabel('Year')
plt.ylabel('Score')
plt.legend()
plt.show()

# USA
plt.figure(figsize=(5,3))
plt.plot(year, usa, 'b-o', label='USA')
plt.title('USA Mathematics Performance')
plt.xlabel('Year')
plt.ylabel('Score')
plt.legend()
plt.show()
```

```python
# 10) Korea, USA, Japan를 한 그래프에 함께 그리기
year = [2006, 2009, 2012, 2015, 2018]
kor  = [547, 546, 554, 524, 526]
usa  = [474, 487, 481, 470, 478]
jpn  = [523, 529, 536, 532, 527]

# 그래프가 가로와 세로 크기를 각각 5인치와 3인치 크기 설정
# year 데이터를 x축으로, kor 데이터를 y축으로 하여 빨간색 점선으로 선 그래프를 그리고 범례에 'Korea'라고 표시

plt.figure(figsize=(5,3))
plt.plot(year, kor, 'r--', label='Korea')
plt.plot(year, usa, 'bo-', label='USA')
plt.plot(year, jpn, 'g-.', label='Japan')

plt.ylim(400, 600)
plt.title('Mathematics Performance Comparison')
plt.xlabel('Year')
plt.ylabel('Score')
plt.legend()
plt.grid(True, linestyle=':', alpha=0.5)
plt.show()
```

```python
# 11) 딕셔너리 데이터로 막대 그래프
notebook = {'g○○':150, 't○○':100, 'm○○':160, 's○○':170}

plt.figure(figsize=(5,3))
# x축: 제품 이름, y축: 가격, 너비(width)=0.7
plt.bar(notebook.keys(), notebook.values(), width=0.7)
plt.xlabel('product name')
plt.ylabel('price')
plt.title('Notebook Prices')
plt.show()
```

```python
# 12) 1인 가구 비율 vs 출생아 수 산점도 + 연도 주석
year = [2000, 2005, 2010, 2015, 2020]
single_person = [15.5, 20.0, 23.9, 27.2, 31.7]
baby_born = [640089, 438707, 470171, 438420, 272337]

plt.figure(figsize=(5,3))
# 산점도
plt.scatter(single_person, baby_born)
# 각 점 옆에 연도 텍스트 추가
for i in range(len(year)):
    plt.annotate(year[i],
                 xy=(single_person[i], baby_born[i]),
                 xytext=(5, -10),
                 textcoords='offset points')
# x,y축 범위 설정
plt.xlim(14, 35)
plt.ylim(230000, 700000)
plt.xlabel('single-person household ratio (%)')
plt.ylabel('the number of babies born')
plt.title('Single-Person Ratio vs Birth Count')
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

# 13) 히스토그램: 1~100 난수 1000개, 20개 구간, 투명도 0.5
x = [random.randint(1, 100) for i in range(1000)]

plt.figure(figsize=(5,3))
plt.hist(x, bins=20, alpha=0.5, edgecolor='black')
plt.ylabel('frequency')
plt.title('Random Number Distribution')
plt.show()
```

```python
# 14) 사용자 입력 기반 막대 그래프
exercise = {'climbing':0, 'soccer':0, 'swimming':0, 'tennis':0}

print('좋아하는 운동을 입력하세요.')
print('복수 개의 운동을 입력하려면 콤마로 구분하세요.')

while True:
    s = input('climbing,soccer,swimming,tennis,종료: ')
    if s == '종료':
        break
    for a in s.split(','):
        a = a.strip()
        if a in exercise:
            # 해당 운동의 카운트를 1 증가
            exercise[a] += 1
        else:
            print(f"'{a}'는 등록되지 않은 운동입니다.")

plt.figure(figsize=(5,3))
plt.bar(exercise.keys(), exercise.values(), color='g', width=0.6)
plt.xlabel('exercise')
plt.ylabel('frequency')
plt.title('Favorite Sports Frequency')
plt.show()
```
