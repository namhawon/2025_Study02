## 문제 2: 회문 판별하기

```python
s1 = input('문자열: ')
s2 = ""

for i in range(len(s1)-1, -1, -1):
    s2 = s2 + s1[i]

if s1 == s2:
    print('회문')
else:
    print('회문아님')
```

# 2. 문자열 다루기

## 문제 1: 대소문자 변환하기

```python
lower_list = []  # 소문자 단어들을 저장할 리스트

while True:
    s = input('영어 단어 또는 exit: ')  # 사용자 입력 받기
    if s == 'exit':  # 'exit' 입력 시 종료
        break
    lower_list.append(s.lower())  # 소문자로 변환해서 리스트에 추가

for a in lower_list:  # 저장된 단어들 출력
    print(a)
```

## 문제 4: 문자열 분리와 결학하기

## 부산시 금정구

```python
# 문자열 → 리스트 (split 사용)
s = "부산시, 금정구"
print("원본 문자열:", s)

# 쉼표(,) 기준으로 문자열을 나눔 → 리스트로 반환
words = s.split(', ')
print("split 결과 (리스트):", words)
```

```python
# 리스트 → 문자열 (join 사용)
# 리스트의 요소들을 쉼표(,)로 연결하여 하나의 문자열로 만듦
joined = ', '.join(words)
print("join 결과 (문자열):", joined)
```

- **공백 기준 분리**

```python
# 문자열 → 리스트 (공백 기준 분리)
s = "부산시 금정구"
print("원본 문자열:", s)

# split()의 기본 동작은 공백 기준 분리
words = s.split()
print("split 결과 (리스트):", words)
```

## 문제 5: 문자열 구성 파악하기

```python
while True:
    id = input('아이디: ')
    if len(id)<5:
        print('길이가 짧습니다.')
    elif id.isalnum()!=True:
        print('아이디는 알파벳과 숫자로 이루어져야 합니다.')
    else:
        break
print(f'아이디는 {id}입니다.')
```

## 문제 5: 형 변환하기

```python
s = input('키: ')

if s.find('cm') != -1:
    height = float(s.rstrip('cm'))
else:
    height = float(s)

print(height)
```
