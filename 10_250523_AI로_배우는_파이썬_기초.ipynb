## 9. 문제 9: 이상치 데이터 탐색 및 제거
- 데이터프레임에서 점수와 공부시간의 이상치를 처리하세요.
- 사분위수를 이용한 판별과 제거

```python
import pandas as pd

# 예: df는 기존 데이터프레임

def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df[(df[column] >= lower) & (df[column] <= upper)]

# 점수에서 이상치 제거
df_no_outliers_score = remove_outliers_iqr(df, '점수')

# 공부시간에서 이상치 제거 (점수 이상치 제거 후 적용)
df_no_outliers_both = remove_outliers_iqr(df_no_outliers_score, '공부시간')

# 결과 출력
print("이상치가 제거된 데이터프레임:")
print(df_no_outliers_both)
```

## 11. 문제 11: 데이터 시각화

### **11-1. 각 '분반'의 평균 점수를 막대 그래프로 시각화하세요**


```python
import pandas as pd
import matplotlib.pyplot as plt

# 분반별 점수 평균 계산
grouped = df.groupby('분반')['점수'].mean()

# 막대그래프 그리기
plt.figure(figsize=(8, 5))
grouped.plot(kind='bar', color='skyblue')
plt.title('분반별 평균 점수')
plt.xlabel('분반')
plt.ylabel('평균 점수')
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import seaborn as sns

# 집계 데이터프레임으로 변환
grouped_df = df.groupby('분반', as_index=False)['점수'].mean()

# 시각화
plt.figure(figsize=(8, 5))
sns.barplot(data=grouped_df, x='분반', y='점수', color='skyblue')
plt.title('분반별 평균 점수')
plt.xlabel('분반')
plt.ylabel('평균 점수')
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
